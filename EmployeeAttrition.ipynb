```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Import data
dataset = pd.read_csv("Attrition data.csv")
```

```python
#Understaanding Data
plt.figure(figsize=(4,3))
s = dataset['Attrition'].value_counts(ascending=True)
sns.countplot(x=dataset['Attrition'], palette='pastel', width = 0.5, hue=None, order=s.index)
plt.title("Overall Attrition in company")
plt.xlabel("Employee Attrition(Yes/No?)")
plt.ylabel("Count")

num_of_employees_left = (dataset["Attrition"]=="Yes").sum()
# print(num_of_employees_left)
attri_rate = num_of_employees_left / len(dataset) * 100

print('\n\n\n')
print(f"Overall Atrrition rate of company XYZ is {round(attri_rate,2)}%")
print(f"i.e.; {num_of_employees_left} employees out of {len(dataset)} have left the company.")
print('\n\n')
```

```python
#Calculating Attrition rate for different sub-groups
plt.figure(figsize=(4,3))
s = dataset['Attrition'].value_counts(ascending=True)
sns.countplot(x=dataset['Attrition'], palette='pastel', width = 0.5, hue=dataset['Gender'], order=s.index)
plt.title("Overall Attrition in company")
plt.xlabel("Employee Attrition(Yes/No?)")
plt.ylabel("Count")

num_of_female_employees_left = ((dataset['Attrition'] == 'Yes') & (dataset['Gender'] == 'Female')).sum()
total_female_employees = (dataset['Gender'] == 'Female').sum()
attri_rate_female = num_of_female_employees_left / total_female_employees * 100

num_of_male_employees_left = ((dataset['Attrition'] == 'Yes') & (dataset['Gender'] == 'Male')).sum()
total_male_employees = (dataset['Gender'] == 'Male').sum()
attri_rate_male = num_of_male_employees_left / total_male_employees * 100

print(f"Attrition rate among females is {round(attri_rate_female, 2)}%")
print(f"Attrition rate among males is {round(attri_rate_male, 2)}%")
print('\n\n')
```

```python
plt.figure(figsize=(8,6))
s = dataset['Attrition'].value_counts(ascending=True)
sns.countplot(x=dataset['Attrition'], palette='pastel', width = 0.5, hue=dataset['Department'], order=s.index)
plt.title("Overall Attrition in company")
plt.xlabel("Employee Attrition(Yes/No?)")
plt.ylabel("Count")

num_of_sales_employees_left = ((dataset['Attrition'] == 'Yes') & (dataset['Department'] == 'Sales')).sum()
total_sales_employees = (dataset['Department'] == 'Sales').sum()
attri_rate_sales = num_of_sales_employees_left / total_sales_employees * 100

num_of_research_employees_left = ((dataset['Attrition'] == 'Yes') & (dataset['Department'] == 'Research & Development')).sum()
total_research_employees = (dataset['Department'] == 'Research & Development').sum()
attri_rate_research = num_of_research_employees_left / total_research_employees * 100

num_of_human_employees_left = ((dataset['Attrition'] == 'Yes') & (dataset['Department'] == 'Human Resources')).sum()
total_human_employees = (dataset['Department'] == 'Human Resources').sum()
attri_rate_human = num_of_human_employees_left / total_human_employees * 100

print(f"Attrition rate among Sales Department is {round(attri_rate_sales, 2)}%")
print(f"Attrition rate among Research & Development Department is {round(attri_rate_research, 2)}%")
print(f"Attrition rate among Human Resources Department is {round(attri_rate_human, 2)}%")
print('\n\n')
```

```python
plt.figure(figsize=(8,6))
s = dataset['Attrition'].value_counts(ascending=True)
sns.countplot(x=dataset['Attrition'], palette='pastel', width = 0.5, hue=dataset['JobRole'], order=s.index)
plt.title("Overall Attrition in company")
plt.xlabel("Employee Attrition(Yes/No?)")
plt.ylabel("Count")

top_attri_job_roles = ['Research Scientist', 'Sales Executive', 'Laboratory Technician']
print(f'The job roles with highest attrition rate are - {top_attri_job_roles}\n\n')

for job_role in top_attri_job_roles:
  num_of_job_employees_left = ((dataset['Attrition'] == 'Yes') & (dataset['JobRole'] == job_role)).sum()
  total_job_employees = (dataset['JobRole'] == job_role).sum()
  attri_rate_job = num_of_job_employees_left / total_job_employees * 100
  print(f"Attrition rate among {job_role} role is {round(attri_rate_job, 2)}%")
print('\n\n')
```

```python
#Building a Correlation matrix
dataset = pd.read_csv('Attrition data.csv')
dataset['Attrition'] = dataset['Attrition'].map({'Yes': 1, 'No': 0})

#One-Hot Encode the non-numeric columns
dataset_encoded = pd.get_dummies(dataset, drop_first=True)

corr_matrix = dataset_encoded.corr()

plt.figure(figsize=(25,25))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Getting insights form the Correlation Matrix

plt.figure(figsize=(15, 8))

# Boxplot for numeric variables
plt.subplot(2, 3, 1)
sns.boxplot(x='Attrition', y='Age', data=dataset, palette='pastel')
plt.title('Attrition by Age')

plt.subplot(2, 3, 2)
sns.boxplot(x='Attrition', y='YearsWithCurrManager', data=dataset, palette='pastel')
plt.title('Attrition by Years With Current Manager')

plt.subplot(2, 3, 3)
sns.boxplot(x='Attrition', y='JobSatisfaction', data=dataset, palette='pastel')
plt.title('Attrition by Job Satisfaction')

# Count plot for categorical variables
plt.subplot(2, 3, 4)
sns.countplot(x='Attrition', hue='BusinessTravel', data=dataset, palette='pastel')
plt.title('Attrition by Business Travel')

plt.subplot(2, 3, 5)
sns.countplot(x='Attrition', hue='MaritalStatus', data=dataset, palette='pastel')
plt.title('Attrition by Marital Status')

plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

#Predictive Analysis
numerical_cols = dataset_encoded.select_dtypes(include='number').columns
dataset_encoded[numerical_cols] = dataset_encoded[numerical_cols].fillna(dataset_encoded[numerical_cols].mean())

# Remove rows with missing values in categorical features
categorical_cols = dataset_encoded.select_dtypes(exclude='number').columns
dataset_encoded.dropna(subset=categorical_cols, inplace=True)
X = dataset_encoded.drop(columns=['Attrition', 'EmployeeID'])
y = dataset_encoded['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=108, test_size=0.3)

# Train logistic regression model
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Inspecting False Positive cases

results_df = pd.DataFrame({'True Label': y_test, 'Predicted Label': y_pred})
false_positives = results_df[(results_df['True Label'] == 0) & (results_df['Predicted Label'] == 1)]

print("Few Employees are at risk of leaving the company as predicted by the model. Their Employee IDs are as follows-")
print(false_positives.index.sort_values().tolist())
```
